## Waste Separation

### Large-Economic-Units-that-Perform-waste-separation

```python
# Lectura del archivo csv
root ="/content/drive/MyDrive/Universidad/4to Semestre/Periodo 3/Ethics/Phase3/Waste Separation/Large-Economic-Units-that-Perform-waste-separation.csv"
df_LEUTPWS = pd.read_csv(root)
```

```python
df_LEUTPWS.shape # Dimensiones del dataframe
```

```python
df_LEUTPWS.head(5) # Primeras 5 filas del dataframe
```

```python
len(df_LEUTPWS["Environmental Process"])
```

```python
colors = ["LIMEGREEN","ORANGE"]

# explosion
explode = []
for i in range(len(df_LEUTPWS["Environmental Process"])):
  explode.append(0.01)

print(explode)
  
# Pie Chart
values = df_LEUTPWS["Value"]
lbls = df_LEUTPWS["Environmental Process"]

plt.pie(values, labels=values,colors=colors,
        autopct='%1.1f%%', pctdistance=0.85,
        explode=explode)
  
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(centre_circle)
  
# Adding Title of chart
plt.title('Large Economic Units that Perform waste separation')

plt.legend(lbls, bbox_to_anchor=(0.5,0.05)) 
# Displaying Chart
plt.show()
```

### Large-Economic-Units-According-to-waste-separation

```python
# Lectura del archivo csv
root = "/content/drive/MyDrive/Universidad/4to Semestre/Periodo 3/Ethics/Phase3/Waste Separation/Large-Economic-Units-According-to-waste-separation.csv"
df_LEUATWS = pd.read_csv(root)
```

```python
df_LEUATWS.head(5)
```

```python
# Figure Size
fig = plt.figure(figsize =(15, 7))

x = df_LEUATWS["Environmental Process"]
y = df_LEUATWS["Value"]
percentage = df_LEUATWS["Percentage"]

# Colours
col_map = plt.get_cmap('Paired')

# Horizontal Bar Plot
graph = plt.bar(x, y,color=col_map.colors)

# Labels
plt.xlabel(x.name)
plt.ylabel(y.name)
plt.legend()
plt.grid(axis="y")
plt.title("Percentage of Large Economic Units According to waste separation")

# Show Percentage
i = 0
for p in graph:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    plt.text(x+width/2,
             y+height*1.01,
             str(round(df_LEUATWS["Percentage"][i],2))+'%',
             ha='center',
             weight='bold')
    i+=1
# Show Plot
plt.show()
```
